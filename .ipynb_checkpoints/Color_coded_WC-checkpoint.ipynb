```python
color_map = [x for x in cm.get_cmap('tab20b').colors]
color_map.extend([x for x in cm.get_cmap('tab20c').colors])
hex_colors = [rgb2hex(x) for x in color_map]
#looking at the tab20 colors, i put in order and by theme
cn2cmap_index = {
    'neutral' : [9],
    'cool' : [8, 5, 0],
    'hot' : [4, 3, 6],
    'forest' : [2, 1, 7],
}
cmap_index2cn = {x: k for k, v in cn2cmap_index.items() for x in v}
print(len(cmap_index2cn))
cmap_cns = [(hex_colors[color_idx*4+shade_idx], cmap_index2cn[color_idx]+'_'+str(color_idx)+'_'+str(shade_idx)) for color_idx in cmap_index2cn for shade_idx in range(4)]
CN2CMAP = {cn:cmap for cmap, cn in cmap_cns}
CMAP2CN = {v:k for k, v in CN2CMAP.items()}
df = pd.DataFrame([{'y axis': 10, 'color': cm, "label":l} for l, cm in CN2CMAP.items()])

print("length", len(df))
# display contents of dataframe
df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

import json
import random 

K = 15
```

```python
word_freqs = {'word_'+str(i):i+2 for i in range(5)}

cat_colors = {
    "Forest" : ["#637939", "#8c6d31" ,"#31a354" ,"#e7ba52" ,"#b5cf6b" ,"#a1d99b" ,"#969696"],
    "Cool" : ["#636363", "#756bb1", "#3182bd", "#393b79", "#9ecae1", "#6b6ecf"],
    "Heat" : ["#7b4173", "#843c39", "#e6550d", "#d6616b", "#fdae6b"]
}
#very weird comprehension dictionary with fake words and their frequencies per category
C2W_DICOS = { category : { x : {"word_"+str(cat_colors[category].index(x))+'_'+str(i):random.randint(1, 20) for i in range(K)} for x in cat_colors[category]} for category in cat_colors}

    

class SimpleGroupedColorFunc(object):
    """This function is taken form WordCloud documentation
    
    https://amueller.github.io/word_cloud/auto_examples/colored_by_group.html
    
    
    Create a color function object which assigns EXACT colors
       to certain words based on the color to words mapping

       Parameters
       ----------
       color_to_words : dict(str -> list(str))
         A dictionary that maps a color to the list of words.

       default_color : str
         Color that will be assigned to a word that's not a member
         of any value from color_to_words.
    """

    def __init__(self, color_to_words, default_color):
        self.word_to_color = {word: color
                              for (color, words) in color_to_words.items()
                              for word in words}

        self.default_color = default_color

    def __call__(self, word, **kwargs):
        return self.word_to_color.get(word, self.default_color)    
    
    
    
def plot_cloud_from_freq(freq_dico, grouped_color_func,  bn, c_dico):    
    wc = WordCloud( height = 1000, 
                          width = 1000, #max_font_size=50, 
                          max_words=500,
                          background_color="white").generate_from_frequencies(freq_dico) # background_color="white"
    wc.recolor(color_func = grouped_color_func)
    plt.figure(figsize= (20,20))
    plt.tight_layout(pad=0)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    
    custom_lines = [Line2D([0], [0], color=c, lw=15) for c in c_dico.keys()]
    
    plt.legend(custom_lines, [CMAP2CN[c] for c in c_dico.keys()], bbox_to_anchor=(1,1), loc="upper left")
    plt.savefig('{}.png'.format(bn))


def get_top_k_tfcd(freq_dico, colors_to_words, k):
    # function that takes a freq dico and returns the dictionary with top K terms and neutral counts for Focus group visuals
    top_K = dict()
    for color in colors_to_words:
        freq_dico_per_color = {k:v for k, v in freq_dico.items() if k in colors_to_words[color]}
        sort = sorted(freq_dico_per_color, key = lambda x: freq_dico[x])
        if len(sort) >= k:
            top_K.update({x:i+1 for i, x in enumerate(sort[-k:])})
        else:
            top_K.update({x:i for i, x in enumerate(sort)})
        print(color, " color dico:  \n",{x:i+1 for i, x in enumerate(sort[-k:])}, '\n\n')

    return top_K
    
def main(c2w_ds):
    for bn, cat_c_d in c2w_ds.items():
        top_k_dico = dict()
        for c_d in cat_c_d:
            top_k_dico.update(cat_c_d[c_d])
        default_color = 'grey'
        
        grouped_color_func = SimpleGroupedColorFunc({k:v.keys() for k, v in cat_c_d.items()}, default_color)
        plot_cloud_from_freq(top_k_dico, grouped_color_func, bn, cat_c_d)
        
    
        
        
        
```
